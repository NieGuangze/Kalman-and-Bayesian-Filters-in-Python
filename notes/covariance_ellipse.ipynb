# Covariance Ellipse

## Covariance Coefficiency

$$
\rho = \frac {cov(X,Y)} {\sqrt{D(X)D(Y)}} = \frac {E(XY) - E(X)E(Y)} {\sqrt{D(X)D(Y)}}
$$

## Variance Matrix

$$
D(X,Y) = 
\begin{bmatrix}
cov(X, X) & cov(X, Y) \\
cov(X, Y) & cov(Y, Y)
\end{bmatrix} = 
\begin{bmatrix}
D(X) & cov(X, Y) \\
cov(X, Y) & D(Y)
\end{bmatrix}
$$

```python
%matplotlib inline
```

```python
from filterpy.stats import plot_covariance_ellipse
import matplotlib.pyplot as plt
import math
import numpy as np

d1=2
d2=6
u1= 2
u2 = 7
rho = .5
cov = rho * math.sqrt(d1*d2)

P = [[d1, cov], [cov, d2]]
plot_covariance_ellipse((u1, u2), P, fc='g', alpha=0.2, 
                        std=[1, 2, 3],
                        title=f'|{d1:.2f} {cov:.2f}|\n|{cov:.2f} {d2:.2f}|')
plt.grid()

k = math.sqrt(d2)/math.sqrt(d1)
x = np.arange(-5,5,.1)
y = (x - u1)*k + u2
y_ = (x - u1) * (-k) + u2
plt.plot(x, y, ':k')

plt.plot(x, y_, ':r')

plt.plot([u1-math.sqrt(d1), u1-math.sqrt(d1)], [-5, 15], 'r')
plt.plot([-5, 5], [u2-math.sqrt(d2), u2-math.sqrt(d2)], 'b')
```
